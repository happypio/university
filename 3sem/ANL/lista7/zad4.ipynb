```python
import pylab
import math
def plot(xs, ys):
    pylab.plot(xs,ys)
    pylab.show()
```

```python
def pn(xs, x):
    ans = 1
    for xi in xs:
        ans *= (x - xi)
    return ans
def rownoodlegle(n):
    h = 2 / ( n - 1)
    points = [(-1 + i*h) for i in range(n)]
    return points
def czebyszewa(n):
    points = [math.cos ( ((2*k + 1) / (2*(n-1) + 2)) * math.pi ) for k in range(n)]
    return points
```

```python
def check(N):
    X = [i/100 for i in range(-100,100)]
    print("Wezly rownoodlegle")
    xs = rownoodlegle(N)
    #print(xs)
    ys = [pn(xs, x) for x in X]
    plot(X,ys)
    print("Wezly czebyszewa")
    xs = czebyszewa(N)
    #print(xs)
    ys = [pn(xs, x) for x in X]
    plot(X,ys)
```

```python
check(5)
```

```python
check(10)
```

```python
check(20)
```
