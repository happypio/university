```python
sentences = []
tests = []
with open('sentences.txt') as f:
    sentences = [i for i in f.read().split() if i != '\n']
with open('test.txt') as f:
    tests = [i for i in f.read().split() if i != '\n']
```

```python
from tqdm.notebook import tqdm
def gen_dict(d, words, n):
    if len(words) == 0:
        if n not in d:
            d[n] = 0
        d[n] += 1
    else:
        if words[0] not in d:
            d[words[0]] = dict()
        gen_dict(d[words[0]] , words[1:], n)
    
def loadNgrams(data):
    Ngrams = [dict() for i in range(6)]
    for gram in tqdm(range(6)):
        for sen in data:
            if len(sen) > gram:
                words = [sen[i] for i in range(gram)]
                for nxt in sen[gram:]:
                    gen_dict(Ngrams[gram], words, nxt)
                    for i in range(len(words)):
                        if i == len(words) - 1:
                            words[i] = nxt
                        else:
                            words[i] = words[i + 1]
    return Ngrams
                    
    
```

```python
Ngrams = loadNgrams(sentences)
```

```python
from math import log as ln
def check_dict(d, words, n):
    if len(words) == 0:
        if n in d:
            return d[n]
        return False
    else:
        if words[0] in d:
            return check_dict(d[words[0]], words[1:], n)
        return False
        
def count_probability(w):
    p = 0
    if len(w) > 5:
        words = [w[i] for i in range(5)]
        for n in w[4:]:
            c6 = check_dict(Ngrams[5], words, n)
            c5 = check_dict(Ngrams[4], words[:4], words[4])
            c4 = check_dict(Ngrams[3], words[:3], words[3])
            c3 = check_dict(Ngrams[2], words[:2], words[2])
            c2 = check_dict(Ngrams[1], words[:1], words[1])
            if c5 and c6:
                p += ln(0.6 * (c6 / c5) + 0.2 * (c5/c4) + 0.1 * (c4/c3) + 0.1 * (c3/c2))
            else:
                if c4 and c5:
                    p += ln(0.2 * (c5/c4) + 0.1 * (c4/c3) + 0.1 * (c3/c2))
                else:
                    if c3 and c4:
                        p += ln(0.1 * (c4/c3) + 0.1 * (c3/c2))
                        
                
        for i in range(len(words)):
            if i == len(words) - 1:
                words[i] = n
            else:
                words[i] = words[i + 1]
    if p != 0:
        print(p, w)
    return p
def check(w1, w2):
    P_w1 = count_probability(w1)
    P_w2 = count_probability(w2)
    #print(w1, P_w1)
    #print(w2, P_w2)
    if (P_w1 > P_w2):
        return w1
    return w2
```

```python
summ = 0
g = 0
for i in range(0, len(tests), 2):
    summ += 1
    g_word = check(tests[i], tests[i + 1])
    if g_word == tests[i]:
        g += 1
print(g/summ)
```
