```python
from tqdm.notebook import tqdm
def gen_dict(d, words, n):
    if len(words) == 0:
        if n not in d:
            d[n] = 0
        d[n] += 1
    else:
        if words[0] not in d:
            d[words[0]] = dict()
        gen_dict(d[words[0]] , words[1:], n)
    
def loadNgrams(data, number):
    Ngrams = [dict() for i in range(number)]
    for gram in tqdm(range(number)):
        for sen in data:
            if len(sen) > gram:
                words = [sen[i] for i in range(gram)]
                for nxt in sen[gram:]:
                    gen_dict(Ngrams[gram], words, nxt)
                    for i in range(len(words)):
                        if i == len(words) - 1:
                            words[i] = nxt
                        else:
                            words[i] = words[i + 1]
    return Ngrams
                    
    
```

```python
Ngrams = loadNgrams(sentences, 8)
```

```python
from math import log as ln
def check_dict(d, words, n):
    if len(words) == 0:
        if n in d:
            return d[n]
        return False
    else:
        if words[0] in d:
            return check_dict(d[words[0]], words[1:], n)
        return False

lambdas = [[1], [1], [0.8, 0.2], [0.7, 0.2, 0.1], [0.6, 0.3, 0.06, 0.04], [0.6, 0.2, 0.1, 0.06, 0.04], [0.6, 0.2, 0.1, 0.05, 0.03, 0.02]] # probs 2 grams, 3 grams, 4 grams, 5 grams, 6 grams
def prob(c, lambds):
    p = 0
    l = 0
    for i in range(len(c)):
        if c[i]:
            l = i + 1
    #print(len(c), l)
    for i in range(len(c) - l, len(lambds)):
        p += lambds[i] * c[-1 - i] / c[-2 - i]
    if p <= 0:
        return -100000
    return ln(p)
            
def count_probability(w):
    p = 0
    length = 0
    for l in range(8):
        if len(w) > l:
            length = l
    words = [w[i] for i in range(length)]
    words.append('')
    for n in w[length:]:
        words[length] = n
        counts = [check_dict(Ngrams[length - i], words[i:length], words[length]) for i in reversed(range(length + 1))]
        p += prob(counts, lambdas[length - 1])
        for i in range(len(words) - 1):
            words[i] = words[i + 1]
    #if p != 0:
        #print(p, w)
    return p
def check(w1, w2):
    P_w1 = count_probability(w1)
    P_w2 = count_probability(w2)
    #print(w1, P_w1)
    #print(w2, P_w2)
    if (P_w1 >= P_w2):
        return w1
    #print(w1, P_w1)
    #print(w2, P_w2)
    return w2
```

```python
summ = 0
g = 0
for i in range(0, len(tests), 2):
    summ += 1
    g_word = check(tests[i], tests[i + 1])
    if g_word == tests[i]:
        g += 1
print(g/summ)
```
