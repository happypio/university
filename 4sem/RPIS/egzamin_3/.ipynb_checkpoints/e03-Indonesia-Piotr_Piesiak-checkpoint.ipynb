```python
import csv
#Date_reported 0 Country_code 1 Country 2 WHO_region 3 New_cases 4
#Cumulative_cases 5 New_deaths 6 Cumulative_deaths 7

#reading data
data = [] 
with open('WHO-COVID-19-global-data.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        if(row[2] == 'Indonesia'):
            data.append(row)
```

## Część 1

```python
#data averaging
for i in range(len(data)):
    #moving average 7 - days
    data[i][4] = round ( sum ( int(data[j][4]) 
        for j in range(i, min(i + 7, len(data)))) / (min(i + 7, len(data)) - i))
    data[i][6] = round ( sum ( int(data[j][6]) 
        for j in range(i, min(i + 7, len(data)))) / (min(i + 7, len(data)) - i))
```

#### Regresja liniowa:
y w zależności od x przybliżamy wzorem y = ax + b , gdzie
$$a = \frac{\sum_{i=1}^{n}x_i\cdot y_i - n\bar{x}\bar{y}}{\sum_{i=1}^{n}x_i^2 - n\bar{x}^2}$$

$$b = \bar{y} - a\bar{x}$$
oznaczenie 
$$\bar{z} =  \frac{\sum^n_{i = 1}z_i}{n}$$

#### Oszacowanie funkcji regresji:
Korzystam z metody odchylenia standardowego składnika resztowego, który informuje o ile rzeczywiste wartości zmiennej odchylają się od wartości przybliżonych (obliczonych) za pomocą funkcji regresji. oczywiście im mniejsza wartość tego miernika, tym lepsza jakość funkcji regresji.
$$S_{y}=\sqrt{\frac{\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^{2}}{n-m-1}}$$
gdzie m oznacza ilość zmiennych użytych do przybliżenia y, w tym przypadku $m = 1$
$$S_{y}=\sqrt{\frac{\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^{2}}{n-2}}$$

```python
import math
import matplotlib.pyplot as plt
import numpy as np

#function to find best intervals for regression
def find_reg(interval, model, mod_beg = 0):
    #best_a, best_b, best_start, best_sy 
    ans = (0,0,0, math.inf)
    for start in range(mod_beg, len(model) - (interval - 1)):
        n = interval
        x = [i + 1 for i in range(start, start + n)]
        y = [model[i] for i in range(start, start + n)]
        sum1 = sum(x[i] * y[i] for i in range(n))
        sum2 = sum(x[i] ** 2 for i in range(n))
    
        avg_x = sum(x) / n
        avg_y = sum(y) / n

        a  = (sum1 - n * avg_x * avg_y) / (sum2 - n * avg_x ** 2)
        b = avg_y - a * avg_x
        
        sum3 = sum( (y[i] - a * x[i] + b) ** 2 for i in range(n))
        s_y  = math.sqrt( (sum3) / (n - 2) )
        
        if (s_y < ans[3]):
            ans = (a, b, start, s_y)
    return ans

#plot
def plot(start, n, model, titles, a, b):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure()
    ax = plt.axes()
    x = [i + 1 for i in range(start, start + n)]
    y = [model[i] for i in range(start, start + n)]
    reg = [a*x[i] + b for i in range(n)]
    ax.plot(x, reg);
    plt.plot(x, y, 'o')
    plt.xlabel(titles[0])
    plt.ylabel(titles[1])
```

### Proste regresji: zachorowania względem daty oraz zgony względem daty

```python
a,b,start,sy = find_reg(501, [i[4] for i in data])
print(f'Zachorowania = {a} * czas_trwania + {b}')
plot(start, len(data), [i[4] for i in data], ['czas_trwania', 'Zachorowania'], a, b)
a,b,start,sy = find_reg(501, [i[6] for i in data])
print(f'Śmierci = {a} * czas_trwania + {b}')
plot(start, len(data), [i[6] for i in data], ['czas_trwania', 'Śmierci'], a, b)
```

## Część 2

#### Współczynnik korelacji Pearsona
W przypadku gdy nie znamy dokładnego rozkładu, stosujemy estymator współczynnika korelacji
$$\rho_{X,Y}={\frac {\sum _{i=1}^{n}(x_{i}-{\overline {x}})(y_{i}-{\overline {y}})}{{\sqrt {\sum _{i=1}^{n}(x_{i}-{\overline {x}})^{2}}}{\sqrt {\sum _{i=1}^{n}(y_{i}-{\overline {y}})^{2}}}}}$$

```python
# two neighbours: Malaysia and Philipines
malaysia = []
philippines = []
with open('WHO-COVID-19-global-data.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        if(row[2] == 'Malaysia'):
            malaysia.append(row)
        if(row[2] == 'Philippines'):
            philippines.append(row)

#data averaging
for i in range(len(malaysia)):
    #moving average 7 - days
    malaysia[i][4] = round ( sum ( int(malaysia[j][4]) 
        for j in range(i, min(i + 7, len(malaysia)))) / (min(i + 7, len(malaysia)) - i))
    
    malaysia[i][6] = round ( sum ( int(malaysia[j][6]) 
        for j in range(i, min(i + 7, len(malaysia)))) / (min(i + 7, len(malaysia)) - i))
    
    philippines[i][4] = round ( sum ( int(philippines[j][4]) 
        for j in range(i, min(i + 7, len(philippines)))) / (min(i + 7, len(philippines)) - i))
    
    philippines[i][6] = round ( sum ( int(philippines[j][6]) 
        for j in range(i, min(i + 7, len(philippines)))) / (min(i + 7, len(philippines)) - i))
```

```python
#function to find best intervals for correlation
def find_corr(interval, model1, model2, mod_beg = 0):
    #best_start, best_corr
    ans = (0, 0)
    for start in range(mod_beg, len(model1) - (interval - 1)):
        n = interval
        x = [model1[i] for i in range(start, start + n)]
        y = [model2[i] for i in range(start, start + n)]
        x_avg = sum(x)/n
        y_avg = sum(y)/n
        
        sum_xy = sum((x[i] - x_avg)*(y[i] - y_avg) for i in range(n))
        sum_x = sum((x[i] - x_avg)**2 for i in range(n))
        sum_y = sum((y[i] - y_avg)**2 for i in range(n))
        
        corr = sum_xy / (math.sqrt(sum_x) * math.sqrt(sum_y))
        if abs(corr) > abs(ans[1]):
            ans = (start,corr)
    return ans

def find_reg2(interval, model1, model2, mod_beg = 0):
    #best_a, best_b, best_start, best_sy 
    ans = (0,0,0, math.inf)
    n = interval
    x = [model1[i] for i in range(start, start + n)]
    y = [model2[i] for i in range(start, start + n)]
    sum1 = sum(x[i] * y[i] for i in range(n))
    sum2 = sum(x[i] ** 2 for i in range(n))
    
    avg_x = sum(x) / n
    avg_y = sum(y) / n

    a  = (sum1 - n * avg_x * avg_y) / (sum2 - n * avg_x ** 2)
    b = avg_y - a * avg_x
        
    sum3 = sum( (y[i] - a * x[i] + b) ** 2 for i in range(n))
    s_y  = math.sqrt( (sum3) / (n - 2) )
        
    if (s_y < ans[3]):
        ans = (a, b, start, s_y)
    return ans

def plot(start, n, model1, model2, titles):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure()
    ax = plt.axes()
    x = [i + 1 for i in range(start, start + n)]
    y1 = [model1[i] for i in range(start, start + n)]
    y2 = [model2[i] for i in range(start, start + n)]
    plt.plot(x, y1, 'o', label = titles[2])
    plt.plot(x, y2, 'o', label = titles[3])
    plt.xlabel(titles[0])
    plt.ylabel(titles[1])
    
def plot2(start, n, model1, model2, a, b):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure()
    ax = plt.axes()
    x = [model1[i] for i in range(start, start + n)]
    y = [model2[i] for i in range(start, start + n)]
    reg = [a*x[i] + b for i in range(n)]
    ax.plot(x, reg);
    plt.plot(x, y, 'o')
```

### Macierz korelacji dla zachorowań

```python
#correlation between Indonesia and Malaysia
cor_im = find_corr(min(len(data),len(malaysia)), [i[4] for i in data], [i[4] for i in malaysia])
#correlation between Philippines and Malaysia
cor_pm = find_corr(min(len(philippines),len(malaysia)), [i[4] for i in philippines], [i[4] for i in malaysia])
#correlation between Indonesia and Philippines
cor_ip = find_corr(min(len(data),len(philippines)), [i[4] for i in data], [i[4] for i in philippines])

print('kraj', 'I', 'M', 'P')
print('I   ', '1.000', round(cor_im[1], 3), round(cor_ip[1], 3))
print('M   ', round(cor_im[1], 3), '1.000', round(cor_pm[1], 3))
print('P   ', round(cor_ip[1], 3), round(cor_pm[1], 3), '1.000')
```

```python
import datetime
# Indonesia and two neighbours: Malaysia and Philipines
indonesia = []
malaysia = []
philippines = []
population = {'I':273523615 , 
              'M': 32365999, 
              'P': 109581078}
beg = datetime.datetime.strptime('2021-01-01', '%Y-%m-%d')
end = datetime.datetime.strptime('2021-05-10', '%Y-%m-%d')

with open('WHO-COVID-19-global-data.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    i = 0
    for row in reader:
        if i != 0:
            d = datetime.datetime.strptime(row[0], '%Y-%m-%d')
            #if (d <= end and d >= beg):
                #print(d)
            if(row[2] == 'Indonesia' and d <= end and d >= beg):
                indonesia.append(row)
            if(row[2] == 'Malaysia' and d <= end and d >= beg):
                malaysia.append(row)
            if(row[2] == 'Philippines' and d <= end and d >= beg):
                philippines.append(row)
        i += 1

#data averaging
for i in range(len(malaysia)):
    #moving average 7 - days
    malaysia[i][4] = round ( sum ( int(malaysia[j][4]) 
        for j in range(i, min(i + 7, len(malaysia)))) / (min(i + 7, len(malaysia)) - i))
    
    philippines[i][4] = round ( sum ( int(philippines[j][4]) 
        for j in range(i, min(i + 7, len(philippines)))) / (min(i + 7, len(philippines)) - i))
    
    indonesia[i][4] = round ( sum ( int(indonesia[j][4]) 
        for j in range(i, min(i + 7, len(indonesia)))) / (min(i + 7, len(indonesia)) - i))
    

        
#new cases per 1000 people
print(len(indonesia), len(malaysia), len(philippines))
for i in range(len(malaysia)):
    indonesia[i][4] = int(indonesia[i][4]) / (population['I'] / 1000)
    philippines[i][4] = int(philippines[i][4]) / (population['P'] / 1000)
    malaysia[i][4] = int (malaysia[i][4]) / (population['M'] / 1000)
```

#### Hipoteza o równości wartości średnich

#### Test istotności: (test T - Welcha)
$$t = \frac {\overline {X}_{1}- \overline {X}_{2}}{\sqrt{\frac {s_{1}^{2}}{n_{1}}+ \frac {s_{2}^{2}}{n_{2}}}} $$
#### Stopnie swobody rozkładu t-Studenta
$$\nu ={\frac {\left({\frac {s_{1}^{2}}{n_{1}}}+{\frac {s_{2}^{2}}{n_{2}}}\right)^{2}}{{\frac {s_{1}^{4}}{n_{1}^{2}\cdot (n_{1}-1)}}+{\frac {s_{2}^{4}}{n_{2}^{2}\cdot (n_{2}-1)}}}}$$


```python
def count_t_v(model1,model2):
    n1 = len(model1)
    n2 = len(model2)
    
    x1_avg = sum(model1) / n1
    x2_avg = sum(model2) / n2
    
    ss1 = sum((model1[i] - x1_avg)**2 for i in range(n1)) / (n1 - 1)
    ss2 = sum((model2[i] - x2_avg)**2 for i in range(n2)) / (n2 - 1)
    
    den = math.sqrt(ss1/n1 + ss2/n2)
    
    t = (x1_avg - x2_avg) / den
    print(t)
    
    v = ((ss1/n1 + ss2/n2) ** 2) / ((ss1 ** 2) / ((n1**2) * (n1 - 1)) + (ss2 ** 2) / ((n2**2) * (n2 - 1)))
    print(round(v))
    return (t,round(v))
```

```python
import scipy.stats
t,v = count_t_v([i[4] for i in malaysia], [i[4] for i in indonesia])
p_value = 2 * min(scipy.stats.t.cdf(t, df=v), 1 - scipy.stats.t.cdf(t, df=v))
print(p_value)
#Zatem na poziomie istotności alfa = 0.05 hipotezę odrzucamy
```

```python
t,v = count_t_v([i[4] for i in malaysia], [i[4] for i in philippines])
p_value = 2 * min(scipy.stats.t.cdf(t, df=v), 1 - scipy.stats.t.cdf(t, df=v))
print(p_value)
#Zatem na poziomie istotności alfa = 0.05 hipotezę odrzucamy
```

#### Hipoteza o równości wartości średnich używając analizy ANOVA

```python
obs_in_group = len(malaysia) # to samo dla indonesia, philippines = 130
global_avg = sum(i[4] for i in indonesia) + sum(i[4] for i in malaysia) + sum(i[4] for i in philippines)
global_avg /= (3 * obs_in_group)

avgs = {'I': sum(i[4] for i in indonesia) / obs_in_group,
        'M': sum(i[4] for i in malaysia) / obs_in_group,
        'P': sum(i[4] for i in philippines) / obs_in_group}
##SSTot = suma po i,j (x_ij - globalna_avg) ** 2
SSTot = 0
for c in [indonesia,malaysia,philippines]:
    for i in c:
        SSTot += (i[4] - global_avg) ** 2

#SSA = ilość_obserwacji_w_grupie * suma (średnia_w grupie - global_avg) ** 2
SSA = obs_in_group * sum((x - global_avg) ** 2 for x in avgs.values())

##SSTot = SSA + SSE
SSE = SSTot - SSA

MSA = SSA / 2 # (num_of_groups - 1) = 3 - 1
MSE = SSE / 387 # num_of_groups * (obs_in_group - 1 ) = 3 * (130 - 1)

t = MSA / MSE
print(t)

p_value = 2 * min(1 - scipy.stats.f.cdf(t, dfn = 2, dfd = 387), scipy.stats.f.cdf(t, dfn = 2, dfd = 387))

print(p_value)

#Zatem na poziomie istotności alfa = 0.05 hipotezę odrzucamy
```
